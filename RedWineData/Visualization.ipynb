```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
Data = pd.read_csv("Cleaned_Red_Data.csv")
```

```python
Data
```

```python
sns.lineplot(x='quality',y='alcohol',data=Data)
plt.title("Relationship Between the Alcohol Content and Qaulity")
plt.ylabel("Alcohol Content")
plt.xlabel("Quality")
```

```python
unique_data = Data[['quality', 'Avg_FixedAcidity']].drop_duplicates()
unique_data = unique_data.sort_values('Avg_FixedAcidity')  # Sort!

plt.figure(figsize=(6,4))
sns.lineplot(x='Avg_FixedAcidity',y='quality',data=unique_data)
plt.title("Relationship Between the FixedAcidity and Quality ")
plt.xlabel("Average Fixed Acidity ")
plt.ylabel("Quality")
plt.grid(True)
```

```python
# Avg Sulphate vs the Qaulity 
unique_data = Data[['quality', 'Avg_Sulphates']].drop_duplicates()
unique_data = unique_data.sort_values('Avg_Sulphates')  # Sort!


sns.barplot(x='quality',y='Avg_Sulphates',data=unique_data)
```
