```python
import pandas as pd
from sklearn.utils import resample
import seaborn as sns 
```

```python
red_Data= pd.read_csv("winequality-red.csv")

red_Data = red_Data.iloc[:, 0].str.split(';', expand=True)
red_Data.columns=[
    'fixed acidity',
    'volatile acidity',
    'citric acid',
    'residual sugar',
    'chlorides',
    'free sulfur dioxide',
    'total sulfur dioxide',
    'density',
    'pH',
    'sulphates',
    'alcohol',
    'quality']

print(red_Data.head())
```

```python
red_Data.drop("volatile acidity",axis=1 ,inplace=True)
red_Data.drop("free sulfur dioxide",axis=1,inplace=True)
red_Data.drop("residual sugar",axis=1,inplace=True)
red_Data.drop("chlorides",axis=1,inplace=True)
```

```python
red_Data['quality'].unique()
```

```python
red_Data['quality']=red_Data['quality'].astype(int)
red_Data['fixed acidity'] = red_Data['fixed acidity'].astype(float)
red_Data['citric acid'] = red_Data['citric acid'].astype(float)
red_Data['total sulfur dioxide'] = red_Data['total sulfur dioxide'].astype(float)
red_Data ['density'] = red_Data['density'].astype(float)
red_Data ['pH'] =red_Data['pH'].astype(float)
red_Data ['sulphates'] = red_Data['sulphates'].astype(float)
red_Data['alcohol'] = red_Data['alcohol'].astype(float)
```

```python
red_Data.info()
```

```python
red_Data.drop("total sulfur dioxide",axis=1,inplace=True)
red_Data.drop("pH",axis=1,inplace=True)
```

```python
red_Data.corr()
```

```python
red_Data.drop('density',axis=1,inplace=True)
```

```python
red_Data.to_csv("Cleaned_Red_Data.csv",index=False)
```

```python
red_Data.drop_duplicates(inplace=True)
```

```python
red_Data ['quality'].value_counts()
```

```python
# Quality data is Imbalanced data 
Concating_terms = []
n_samples = 500
Unique_Classes = red_Data['quality'].unique()


for Class in Unique_Classes :

    Class_Data = red_Data[red_Data['quality']==Class]
    Current_Length = len(Class_Data)

    if Current_Length < n_samples :
        # then Upsampling 
        resampled_Data = resample (
            Class_Data,
            n_samples=n_samples,
            replace=True,
            random_state= 52
        ) 
    elif Current_Length > n_samples :
        # DownSampling 
        resampled_Data = resample(
            Class_Data,
            n_samples=n_samples,
            replace=False,
            random_state=50
        )
    Concating_terms.append(resampled_Data)

# Quality data is Imbalanced data 
Concating_terms = []
n_samples = 500
Unique_Classes = red_Data['quality'].unique()


for Class in Unique_Classes :

    Class_Data = red_Data[red_Data['quality']==Class]
    Current_Length = len(Class_Data)

    if Current_Length < n_samples :
        # then Upsampling 
        resampled_Data = resample (
            Class_Data,
            n_samples=n_samples,
            replace=True,
            random_state= 52
        ) 
    elif Current_Length > n_samples :
        # DownSampling 
        resampled_Data = resample(
            Class_Data,
            n_samples=n_samples,
            replace=False,
            random_state=50
        )
    Concating_terms.append(resampled_Data)

Balanced_data = pd.concat (Concating_terms,ignore_index=True)
Balanced_data

    
```

```python
Balanced_data.to_csv("Cleaned_Red_Data.csv",index=False)
```

```python
Average_fixed_Acidity = Balanced_data.groupby('quality')['fixed acidity'].mean()
Balanced_data['Avg_FixedAcidity'] = Balanced_data['quality'].map(Average_fixed_Acidity)
```

```python
# Average Sulphates for the different Qaulity 
Average = Balanced_data.groupby('quality')['sulphates'].mean()
Balanced_data['Avg_Sulphates']=Balanced_data['quality'].map(Average)
avg = Balanced_data.groupby ('quality')['volatile acidity'].mean()
Balanced_data['avg_volatile_acidity'] = Balanced_data['quality'].map(avg)
```
