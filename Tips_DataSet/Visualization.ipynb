```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
Data = pd.read_csv('Cleaned_data.csv')
Data
```

```python
sns.scatterplot(x='total_bill',y="tip",data=Data)
plt.title("Relationship B/w Total_bill and tip ")
plt.xlabel("Total Bill")
plt.ylabel("tip")
plt.grid(True)
```

```python
Mean_bill = Data.groupby('day')['total_bill'].mean()
```

```python
Mean_bill.plot(kind='bar')
plt.title('Average Sales')
plt.grid(True)
plt.xlabel("days")
plt.ylabel('amount')
```

```python
sns.barplot(x="total_bill",y="day",data=Data)
plt.title("Relationship B/w Total_bill and day ")
plt.xlabel("Total Bill")
plt.ylabel("day")
plt.grid(True)
```
