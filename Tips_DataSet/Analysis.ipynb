```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.utils import resample
```

```python
Data = sns.load_dataset('tips')
Data
```

```python
Data.isnull().sum()
```

```python
Data['sex'].value_counts()
```

```python
# upsampling the Female Category 
Majority = Data[Data['sex'] == "Male"]
Minority = Data[Data['sex']== "Female"]

Upsampled_Minority = resample(
    Minority,
    n_samples=len(Majority),
    random_state=25,
    replace=True
)
Upsampled_Minority
```

```python
Final = pd.concat([Upsampled_Minority,Majority])
```

```python
Final['smoker'].value_counts()
```

# Upsmapling the sex and Smoker As Group 

```python
Data['Smoker_Gender'] = Data['sex'].astype(str)+'_'+Data['smoker'].astype(str)

Data['Smoker_Gender'].value_counts()

Male_NO= Data[Data['Smoker_Gender'] == "Male_No"]
Male_Yes= Data[Data['Smoker_Gender'] == "Male_Yes"]
Female_No= Data[Data['Smoker_Gender'] == "Female_No"]
Female_Yes= Data[Data['Smoker_Gender'] == "Female_Yes"]
```

```python
# Upsampling the Data 
Upsampled_Male_NO = resample(
    Male_NO,
    replace=True,
    n_samples=100,
    random_state=22
)
Upsampled_Male_Yes = resample(
    Male_Yes,
    replace=True,
    n_samples=100,
    random_state=22
)
Upsampled_Female_No = resample(
    Female_No,
    replace=True,
    n_samples=100,
    random_state=22
)
Upsampled_Female_Yes = resample(
    Female_Yes,
    replace=True,
    n_samples=100,
    random_state=22
)
```

```python
List = [Upsampled_Female_No,Upsampled_Female_Yes,Upsampled_Male_NO,Upsampled_Male_Yes]


dfs = [df.reset_index(drop=True)  for df in List]
Final_Data = pd.concat(dfs,axis=0)
```

```python
Final_Data['Smoker_Gender'].value_counts()
```

```python
Final_Data['sex'].value_counts()
```

```python
Final_Data['smoker'].value_counts()
```

```python
Final_Data.drop("Grouped",axis=1,inplace=True)
Final_Data.drop("smoker",axis=1,inplace=True)
```

```python
Final_Data.drop("sex",axis=1,inplace=True)
```

```python
Final_Data.to_csv("Cleaned_data.csv")
```
