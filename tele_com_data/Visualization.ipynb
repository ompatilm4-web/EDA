```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
Data =pd.read_csv("Transformed_data.csv")
Data
```

```python
sns.boxplot(x='tenure',y='Services',data=Data)
plt.title("Services vs Tenure Time ")
plt.xlabel("Tenure")
plt.ylabel("Services")
```

```python
# Customers Not Using MultiLines
sns.countplot(x="MultipleLines_No",data=Data)

# Conclusion :Approximately 3,500 customers do not subscribe to the Multiple Lines service,
#  while the majority of customers use
#  Multiple Lines or belong to other phone service categories.
```

```python
Data['MultipleLines_No phone service'].value_counts()
# Approximately 3500 Customers Don't get Phone Services At All 
```

```python
Data['OnlineSecurity_No'].value_counts()
## Aproximately 5436 Customers Using the internet Service Don't get Online Security 
```

```python
Data['OnlineSecurity_No internet service'].value_counts()
# Aproximately 8861 Customers Have Internet Connection 
```

```python
sns.countplot(x='Services',hue='OnlineSecurity_No',data=Data)
plt.title("Customer With Diff Services Not having Security !")
plt.ylabel("Count")
plt.xlabel("Services")
```

```python
sns.boxplot(x="OnlineSecurity_No",y='tenure',data=Data)
plt.title("tenure vs Online Security Subscriptions ")
plt.ylabel("Tenure")
plt.xlabel("Subscribed Status ")
```

```python
pd.crosstab(
    Data['Services'],
    Data['OnlineSecurity_No']
)
```
