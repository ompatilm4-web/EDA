```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.utils import resample 
```

```python
Data.drop("residual sugar",axis=1,inplace=True)
Data.drop("citric acid",axis=1,inplace=True)
Data.drop("free sulfur dioxide",axis=1,inplace=True)
Data.drop("pH",axis=1,inplace=True)
# Data.drop(""axis=1,inplace=True)
```

```python
Data
```

```python
# Agregating the Columns
# Average Volatile Acidity 
Average = Data.groupby('quality')['volatile acidity'].mean()
Data['Avg_volatile_acidity'] = Data['quality'].map(Average)
Data.drop("volatile acidity",axis=1,inplace=True)
```

```python
Data
```

```python
# Average Sulphates 
Avg = Data.groupby ('quality')['sulphates'].mean()
Data['avg_Silphates'] = Data['quality'].map(Avg)
```

```python
Data.corr()
```

```python
Data.drop("fixed acidity",axis=1,inplace=True)
```

```python
Data.drop("sulphates",axis=1,inplace=True)
```

```python
Data
```

```python
columns=["chlorides","total sulfur dioxide","alcohol",'density','Avg_volatile_acidity','avg_Silphates','quality']
Data = Data[columns]
```

```python
Data.corr()
```

```python
Data['quality'].value_counts()
```

```python
# Balancing the Data 
Concating_terms = []
n_samples = 2000
Unique_Class = Data['quality'].unique()

for Class in Unique_Class :

    Class_data = Data[Data['quality']==Class]
    Current_Count = len(Class_data)

    if Current_Count < n_samples :
        # Upsampling the data 
        resampled_Data = resample(
            Class_data,
            n_samples=n_samples,
            replace=True,
            random_state=50
        )

    elif Current_Count > n_samples :
        # DownSampling 
        resampled_Data = resample(
            Class_data,
            n_samples=n_samples,
            replace= False,
            random_state= 50
        )

    else :
        print("Not Every time Error Comes !")

    Concating_terms.append(resampled_Data)

Balanced_data = pd.concat(Concating_terms,ignore_index=True)
```
