```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
Data = pd.read_csv("Cleaned_Data.csv")
Data
```

```python
Numeric_data = Data.select_dtypes(include=['number'])
Corr_Coef = Numeric_data.corr()
```

```python
sns.heatmap(Corr_Coef,annot=True)
```

```python
sns.scatterplot(x="carat",y="price",data=Data)
plt.title("Relationship B/W Carat and Price ")
plt.ylabel("Price")
plt.xlabel("Carat")
plt.grid(True)


# Conclusion Greater the Carat More Expensive tehdiamond Is 
```

```python
sns.histplot(Data['table'],bins=12,kde=True)
plt.title("Distribution of Table ")
plt.ylabel("Count")
plt.xlabel("table Dimension")
plt.grid(True)
```

```python
sns.scatterplot(x="depth",y="price",data=Data)
plt.title("Relationship Between Depth And price  ")
plt.ylabel("Price")
plt.xlabel("Depth")
plt.grid(True)
```

```python
sns.pairplot(Data)
```

```python
sns.boxplot(x='clarity', y='price', data=Data)
plt.title("Clarity vs Price ")
plt.ylabel("Clarity")
plt.xlabel("Price")
plt.grid(True)
```

```python
sns.violinplot(x='Cut_Color', y='price', data=Data)
```

```python
sns.countplot(x='clarity', data=Data)
plt.title("Number of Diamonds per cut Category")
plt.ylabel("No of Diamonds ")
plt.xlabel("Different Cut ")
plt.grid(True)
```
