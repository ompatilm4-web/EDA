# Balancing the Cut Color Column 

```python
import pandas as pd
from sklearn.utils import resample

# Target number of samples for EVERY class
target_n = 2500

# Container to hold our newly balanced subsets
balanced_subsets = []

# Get a list of all unique combinations in Cut_Color
unique_classes = Data['Cut_Color'].unique()

for current_class in unique_classes:
    # 1. Isolate the rows belonging to the current class
    class_data = Data[Data['Cut_Color'] == current_class]
    current_count = len(class_data)
    
    # 2. Determine if we need to upsample (with replacement) or downsample (without replacement)
    if current_count < target_n:
        # Upsampling minority classes (e.g., Fair_J, Good_E, etc.)
        resampled_subset = resample(
            class_data,
            replace=True,      # Mandatory for upsampling
            n_samples=target_n,
            random_state=50
        )
    else:
        # Downsampling majority classes (e.g., Ideal_G, Ideal_E, etc.)
        resampled_subset = resample(
            class_data,
            replace=False,     # Best practice for downsampling to keep rows unique
            n_samples=target_n,
            random_state=50
        )
        
    # Append the balanced subset to our list
    balanced_subsets.append(resampled_subset)

# 3. Combine all balanced subsets back into a single dataframe
Balanced_Data = pd.concat(balanced_subsets).reset_index(drop=True)

# 4. Verify the results
print("--- New Value Counts ---")
print(Balanced_Data['Cut_Color'].value_counts())
```

```python
Balanced_Data.to_csv('Cleaned_Data.csv')
```
