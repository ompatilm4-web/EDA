```python
import pandas as pd 
import numpy as np 
import seaborn as sns 

from sklearn.utils import resample
```

```python
Data = sns.load_dataset('titanic')
```

```python
Data.isnull().sum()
```

```python
Data['age']=Data ['age'].fillna(Data['age'].median())
```

```python
Q1,Q3 = np.quantile(Data['age'],[0.25,0.75])
IQR = Q3 -Q1


median_age = Data['age'].median()


High_fence = Q3 +1.5*(IQR)
Data['age'].max()
Data.loc[Data['age']>High_fence,'age'] = median_age
```

```python
Data['age'].max()
```

```python
Data['embarked']=Data['embarked'].fillna(Data['embarked'].mode()[0])
Data['deck']=Data['deck'].fillna(Data['deck'].mode()[0])
Data['embark_town']=Data['embark_town'].fillna(Data['embark_town'].mode()[0])
```

```python
Data.isnull().sum()
```

```python
Data['sex'].value_counts()
```

```python
# Upsampling Minority Class
majority = Data[Data['sex']=="male"]
minority = Data[Data['sex'] =="female"]
```

```python
Upsampled_Data = resample (
    minority,
    replace=True,
    n_samples=len(majority),
    random_state=50

)
```

```python
Final_data = pd.concat([majority,Upsampled_Data])
```

```python
Final_data['sex'].value_counts()
```

```python
Final_data.isnull().sum()
```

```python
Final_data
```

```python
Final_data['class'].value_counts()
First = Final_data[Final_data['class'] == 'First']
Second = Final_data[Final_data['class'] == 'Second']
Third = Final_data[Final_data['class'] == 'Third']
```

```python
# DownSampling the Data 
DownSampled_Class_First = resample(
    First,
    replace=True,
    n_samples= 200 ,
    random_state=52,
)

DownSampled_Class_Second = resample(
    Second,
    replace=True,
    n_samples= 200 ,
    random_state=52,
)

DownSampled_Class_Third = resample(
    Third,
    replace=True,
    n_samples= 200 ,
    random_state=52,
)
```

```python
Final_data = pd.concat([DownSampled_Class_Third,DownSampled_Class_First,DownSampled_Class_Second]).reset_index(drop=True)
```
