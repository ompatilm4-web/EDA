```python
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt

```

```python
Data = pd.read_csv('Cleaned_Data.csv')
Data
```

```python
sns.lineplot(x='survived',y='age',data=Data)
plt.title("Ages of who Survived ")
plt.xlabel('Survived')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
sns.barplot(x='survived',y='sex',data=Data)
plt.title("Gender of Who Survived ")
plt.xlabel('Survived')
plt.ylabel('Sex')
plt.grid(True)
plt.show()
```
