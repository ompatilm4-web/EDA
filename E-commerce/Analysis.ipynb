```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.utils import resample 
```

```python
Data = pd.read_csv("ecommerce_messy_12300_rows.csv")
Data
```

```python
Data.info()
```

```python
Data.isnull().sum()
```

```python
Mode = Data['product_name'].mode()[0]
Data ['product_name'] =Data['product_name'].fillna(Mode)
```

```python
category_mode =Data['category'].mode()[0]
Data['category'] = Data['category'].fillna(category_mode)
```

```python
sns.histplot(Data['unit_price'],kde=True)
```

```python
Mean_unit_price = Data['unit_price'].mean()
Data['unit_price'] = Data['unit_price'].fillna(Mean_unit_price)
```

```python
mode_city = Data['city'].mode()[0]
Data['city'] = Data['city'].fillna(mode_city)
```

```python
Mode_payment_method = Data['payment_method'].mode()[0]
Data['payment_method']=Data['payment_method'].fillna(Mode_payment_method)
```
